```python
#
import pandas as pd
import numpy as np
import os
from pandas import DataFrame

if not os.path.exists('results'):
    os.makedirs('results')

def write_dataframe(name, df):
    file_path = f"results/{name}.csv"
    print(name)
    print(df.head(10))
    if os.path.exists(file_path):
        print(f"File {name} already exists")
        return
    df.to_csv(file_path, index=True)
    print(f"Saved: {file_path}")


def make_missing_values(
        src_path: str,
        tgt_path: str,
        n_rows: int = 100,
        missing_per_col: float = 0.1,
) -> None:
    """
    This function loads your DataFrame and stores it as a separate .csv file for your experiments with missing data.

    :param src_path: Path to the original source .csv file to read.
    :param tgt_path: Path where the modified .csv file with missing values will be saved.
    :param n_rows: Number of rows to read from the source file.
    :param missing_per_col: The fraction of values to drop per column (e.g., 0.2 for 20% missing data). Must be between 0.0 and 1.0.
    :return: None
    """

    # 1. Load the specific number of rows from the source CSV
    df = pd.read_csv(src_path, nrows=n_rows)

    # 2. Calculate the exact number of missing values to inject per column
    # We use len(df) in case the source file has fewer rows than n_rows
    n_missing = int(len(df) * missing_per_col)

    # 3. Iterate through each column and inject NaNs randomly
    for col in df.columns:
        # Select random indices without replacement, so we don't pick the same row twice
        missing_indices = np.random.choice(df.index, size=n_missing, replace=False)

        # Assign NaN to the selected random indices for the current column
        df.loc[missing_indices, col] = np.nan

    # 4. Save the modified DataFrame to the target path
    df.to_csv(tgt_path, index=False)
```

# Original dataset and adding missing values

```python
financial_regression: DataFrame = pd.read_csv(r"C:\Users\mbv16\Downloads\financial_regression.csv")
write_dataframe("Original", financial_regression)
make_missing_values(r"C:\Users\mbv16\Downloads\financial_regression.csv", r"C:\Users\mbv16\PycharmProjects\pythonProject1\results\financial_with_missing.csv")
financial_regression_with_miss = pd.read_csv(r"C:\Users\mbv16\PycharmProjects\pythonProject1\results\financial_with_missing.csv")
```

# Deleting

```python
# 1. delete rows with miss
def remove_rows_any_na(df):
    return df.dropna(axis=0, how='any')
write_dataframe("delete_rows_with_miss", remove_rows_any_na(financial_regression_with_miss))

# 2. delete cols with miss
def remove_cols_any_na(df):
    return df.dropna(axis=1, how='any')
write_dataframe("delete_cols_with_miss", remove_cols_any_na(financial_regression_with_miss))

# 3. delete rows_by_threshold
def remove_rows_by_threshold(df, threshold=0.3):
    # threshold requires amount of NOT NaN values
    limit = int((1 - threshold) * df.shape[1])
    return df.dropna(axis=0, thresh=limit)
write_dataframe("delete_rows_by_threshold", remove_rows_by_threshold(financial_regression_with_miss))
```

# Front- and backfill

```python
# 4. Back filling
def fill_bfill(df):
    return df.bfill()
write_dataframe("bfill", fill_bfill(financial_regression_with_miss))
# 5. Front filling
def fill_ffill(df):
    return df.ffill()
write_dataframe("ffill", fill_bfill(financial_regression_with_miss))
```

# Numeric methods

```python
# 6.1 Mean
def fill_mean(df):
    df_c = df.copy()
    num_cols = df_c.select_dtypes(include=[np.number]).columns
    df_c[num_cols] = df_c[num_cols].fillna(df_c[num_cols].mean())
    return df_c
write_dataframe("Mean", fill_mean(financial_regression_with_miss))

# 6.2 Median
def fill_median(df):
    df_c = df.copy()
    num_cols = df_c.select_dtypes(include=[np.number]).columns
    df_c[num_cols] = df_c[num_cols].fillna(df_c[num_cols].median())
    return df_c
write_dataframe("Median", fill_median(financial_regression_with_miss))

# 6.3 Мода (Mode - most )
def fill_mode(df):
    df_c = df.copy()
    return df_c.fillna(df_c.mode().iloc[0])
write_dataframe("Mode", fill_mode(financial_regression_with_miss))

# 6.4 indicator for values (-1)
def fill_numeric_indicator(df, value=-1):
    df_c = df.copy()
    num_cols = df_c.select_dtypes(include=[np.number]).columns
    df_c[num_cols] = df_c[num_cols].fillna(value)
    return df_c
write_dataframe("numeric_indicator", fill_numeric_indicator(financial_regression_with_miss))

# 6.5 "missing"
def fill_categorical_indicator(df, value="missing"):
    df_c = df.copy()
    obj_cols = df_c.select_dtypes(exclude=[np.number]).columns
    df_c[obj_cols] = df_c[obj_cols].fillna(value)
    return df_c
write_dataframe("categorical_indicator", fill_categorical_indicator(financial_regression_with_miss))

# 6.6 Interpolation
def fill_linear_interpolation(df):
    df_c = df.copy()
    num_cols = df_c.select_dtypes(include=[np.number]).columns
    df_c[num_cols] = df_c[num_cols].interpolate(method='linear')
    return df_c
write_dataframe("linear_interpolation", fill_linear_interpolation(financial_regression_with_miss))
```

# Before and after analysis

```python
# 1. Col with missing values
target_col = 'sp500 close'

print("--- Statistics BEFORE filling ---")
stats_before = financial_regression_with_miss[target_col].describe()
print(stats_before)

# 2. Apply interpolation method
df_filled_example = financial_regression_with_miss.copy()
df_filled_example[target_col] = df_filled_example[target_col].interpolate(method='linear')

print("\n--- Statistics AFTER filling (Interpolation) ---")
stats_after = df_filled_example[target_col].describe()
print(stats_after)

# SENTENCE EXPLAINING THE CHANGE:
# The mean slightly increased, whereas std slightly decreased because linear interpolation creates smooth transitions between existing data points, thereby reducing the overall data variance.
```

# Final proof

```python
# Clear df
df_final = financial_regression_with_miss.copy()

# 1. delete cols with a lot of missing values(>50%), its more appropriate than trying to fill them
limit = len(df_final) * 0.5
df_final = df_final.dropna(axis=1, thresh=limit)

# 2. fill left miss with interpolation
numeric_cols = df_final.select_dtypes(include=[np.number]).columns
df_final[numeric_cols] = df_final[numeric_cols].interpolate(method='linear')

# 3. fill the very left and right cols
df_final = df_final.ffill().bfill()

# res
print("Final Cleaned DataFrame Head:")
display(df_final.head())

# final check
missing_count = df_final.isna().sum().sum()
print(f"\nFinal check: There are exactly {missing_count} missing values left.")
```

# Concusion

## In this practical work, I explored various strategies for handling missing data using the financial_regression.csv dataset. The key findings are:

## Data Removal: 
While dropna() is straightforward, it significantly reduces the dataset size. Using the thresh parameter proved to be a more balanced approach, allowing the retention of rows that are at least 70% complete.

## Statistical Imputation: 
Filling with mean or median is useful for stable metrics like trading volumes, but it can distort the natural variance of the data.

## Time-Series Methods: 
For financial indicators (S&P 500, Gold, Oil), Linear Interpolation and Forward Fill (ffill) were the most effective. These methods preserve the logical trend and continuity of price movements over time.

## Technical Summary:
I implemented a modular architecture using custom functions for each transformation and a centralized write_dataframe utility for automated saving. The final pipeline combines column filtering (removing columns with >50% missing values) and multi-step filling (interpolation followed by ffill/bfill), resulting in a fully cleaned dataset with zero missing values.
