```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
def print_matrix(matrix):
    for j in range(matrix.shape[0]):
        for i in range(len(matrix[0])):
            print(matrix[j][i], " ", end="")
        print()
```

## I.1.a task

```python
rng = np.random.default_rng()
rand_arr = rng.random(10)
identity_matrix = np.array([[1,0,0],[0,1,0],[0,0,1]])
print(rand_arr)
print_matrix(identity_matrix)
```

## I.1.c task

```python
matrix1 = np.array([[1,2,1],[2,1,2],[1,2,1]])
matrix2 = np.array([[2,1,2],[1,2,1],[2,1,2]])
resulting_matrix_smart = np.arange(9).reshape(3, 3)
resulting_matrix_dot = np.dot(matrix1, matrix2)
for j in range(3):
    for i in range(3):
        resulting_matrix_smart[j][i] = matrix1[j][i] * matrix2[j][i]
print("result smart matrix")
print_matrix(resulting_matrix_smart)
print("result dot matrix")
print_matrix(resulting_matrix_dot)
```
